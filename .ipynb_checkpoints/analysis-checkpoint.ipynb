```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
walk_acc = pd.read_csv('data2/walk/walk_acc.csv', names=['date', 'type', 'x', 'y', 'z'])
```

```python
walk_acc = walk_acc.drop(['type'], axis=1)
```

```python
walk_acc['dur'] = walk_acc['date'] - walk_acc['date'].min()
walk_acc = walk_acc.drop(['date'], axis=1)
```

```python
def plot(name):
    split_name = name.split('_')
    cols = ['x', 'y', 'z']
    if split_name[1] == 'rot':
        cols = ['a', 'b', 'g']
        
    walk_acc = pd.read_csv('data2/{}/{}.csv'.format(split_name[0], name), names=['date', 'type']+cols)
    walk_acc = walk_acc.drop(['type'], axis=1)
    walk_acc['dur'] = walk_acc['date'] - walk_acc['date'].min()
    walk_acc = walk_acc.drop(['date'], axis=1)
    ax = sns.lineplot(x='dur', y='value', hue='variable', data=pd.melt(walk_acc, ['dur']), alpha=0.5)
    ax.set_title('{} in {}'.format(split_name[0].title(), split_name[1].title()))
    ax.set_xlabel('duration in millisec')
    #ax.plot()
    plt.show()
```

```python
data = 'walk'
for i in ['acc', 'gacc', 'rot']:
    plot('{}_{}'.format(data, i))
```

```python
data = 'run'
for i in ['acc', 'gacc', 'rot']:
    plot('{}_{}'.format(data, i))
```

```python
data = 'skip'
for i in ['acc', 'gacc', 'rot']:
    plot('{}_{}'.format(data, i))
```

```python
skip_gacc = pd.read_csv('data2/skip/skip_gacc.csv', names=['date', 'type', 'x', 'y', 'z'])
skip_gacc = skip_gacc.drop(['type'], axis=1)
skip_gacc['dur'] = skip_gacc['date'] - skip_gacc['date'].min()
skip_gacc = skip_gacc.drop(['date'], axis=1)
```

```python
windowed = skip_gacc.rolling(10).mean()
```
