```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
import lightgbm as lgb
from sklearn.model_selection import KFold, cross_val_score
from sklearn.metrics import *
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def toData(path): # 'data2/walk/walk_acc.csv'
    walk_acc = pd.read_csv(path, names=['date', 'type', 'x', 'y', 'z'])
    walk_acc = walk_acc.drop(['type'], axis=1)
    walk_acc['dur'] = walk_acc['date'] - walk_acc['date'].min()
    walk_acc = walk_acc.drop(['date'], axis=1)
    return walk_acc
```

```python
def toWindows(data, start, end, window_length):
    windows = []
    time_now = start

    while time_now < end - window_length:
        window = data.query("@time_now <= dur < @time_now + @window_length")
        windows.append(window)
        time_now = window.iat[0, -1] +1

    return windows
```

```python
def makeStaticData(windows):
    all_data = []
    for w in windows:
        all_data.append(pd.concat([w.mean()[:-1], w.min()[:-1], w.max()[:-1], w.std()[:-1]]))
        #all_data.append(pd.concat([w.mean()[:-1], w.min()[:-1]]))
        #all_data.append(pd.concat([w.mean()[:-1]])) #0.931
        #all_data.append(pd.concat([w.min()[:-1]]))
        #all_data.append(pd.concat([w.max()[:-1]]))
        #all_data.append(pd.concat([w.std()[:-1]]))
        #all_data.append(pd.concat([w.min()[:-2]]))
        
    return pd.concat(all_data, axis=1).T
```

```python
walk_data = makeStaticData(toWindows(toData('data2/walk/walk_acc.csv'), 6000, 74000, 3000))
```

```python
run_data = makeStaticData(toWindows(toData('data2/run/run_acc.csv'), 6000, 74000, 3000))
```

```python
run_data
```

```python
skip_data = makeStaticData(toWindows(toData('data2/skip/skip_acc.csv'), 10000, 94000, 3000))
```

```python
skip_data
```

```python
walk_data['label'] = 'walk'
run_data['label'] = 'run'
skip_data['label'] = 'skip'
data = pd.concat([walk_data, run_data, skip_data])
data.columns = ['a_x', 'a_y', 'a_z', 'i_x', 'i_y', 'i_z', 'x_x', 'x_y', 'x_z', 's_x', 's_y', 's_z' ,'label']
```

```python
#data
```

```python
y = np.array(data['label'])
X = np.array(data.drop(['label'], axis=1))
```

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=7)
```

```python
clf = DecisionTreeClassifier(max_depth=None,random_state=8)
```

```python
scores=[]
s = np.zeros([3,3])
for train,test in kfold.split(X,y):
    clf.fit(X[train], y[train])
    pred = clf.predict(X[test])
    score = accuracy_score(y[test],pred)
    scores.append(score)
    s += np.array(confusion_matrix(y[test],pred))
    
print(scores)
print(sum(scores)/10)
print(s)
```

```python
clf = RandomForestClassifier(n_estimators=382, max_depth=None,random_state=8)
```
